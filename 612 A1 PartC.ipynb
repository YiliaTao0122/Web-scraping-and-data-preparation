```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('/content/drive/MyDrive/612 course/612 assessment 1/Mecca_skincare.csv')
df2=pd.read_csv('/content/drive/MyDrive/612 course/612 assessment 1/Lookfantastic_skincare.csv')
df3=pd.read_csv('/content/drive/MyDrive/612 course/612 assessment 1/Sephora_skincare.csv')
```

```python
# Handle missing values before converting to integer
df1['rating'] = df1['rating'].fillna(0).astype(int)
df2['rating'] = df2['rating'].fillna(0).astype(int)
df3['rating'] = df3['rating'].fillna(0).astype(int)
```

```python
df1['price']=df1['price'].astype(float)
df2['price']=df2['price'].astype(float)
df3['price']=df3['price'].astype(float)
```

```python
df1.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
df3.head()
```

```python
#concatenate the dataframes
merged_df = pd.concat([df1, df2, df3], ignore_index=True)
merged_df
```

```python
print(merged_df.columns)
```

```python
df=merged_df.copy()
print(df.columns)
```

```python
df.isnull().sum()
```

```python
# Find rows with missing values (NaN)
rows_with_missing_values = merged_df[merged_df.isnull().any(axis=1)]
rows_with_missing_values
```

```python
# Add a new column 'product_id' and set it as index
df['product_id'] = range(1, len(df) + 1)
df.set_index('product_id', inplace=True)
df.head(3)
```

```python
# Reset index to default integer-based index
df.reset_index(drop=True, inplace=True)
df.head(3)
```

```python
# Create a new DataFrame by selecting rows based on a condition
# Select rows where price is greater than 50
new_df = df[df['price'] > 50]
print(new_df)
```

```python
# Verify the column names in your DataFrame
print(df.columns)
```

```python
# Set multiple columns as the index
df.set_index(['product_name', 'function'], inplace=True)
df.head(3)
```

```python
# find a product
product = df.loc[('Summer Fridays')]
print(product)
```

```python
# find a specific product
product_name = 'Summer Fridays'
function_name = 'Dream Lip Oil'
product = df.loc[(product_name, function_name)]
print(product)
```

```python
# Sorting in ascending order based on 'price'
sorted_df_asc = df.sort_values(by='price', ascending=True)
sorted_df_asc.head()
```

```python
# Sorting in descending order based on 'rating'
sorted_df_desc = df.sort_values(by='rating', ascending=False)
sorted_df_desc.head()
```

```python
df.describe()
```
