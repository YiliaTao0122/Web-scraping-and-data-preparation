```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.lookfantastic.com/health-beauty/face/skincare-products.list'
```

```python
respond = requests.get(url)
print(respond)
```

```python
response = respond.content
```

```python
soup= BeautifulSoup(response, 'html.parser')
soup
```

```python
table=soup.find('div',{'class':'productListProducts'})
table
```

```python
products = table.find_all('li',class_='productListProducts_product')
products
```

```python
skincare=[]
for product in products:
    # Extracting product name
    product_name = product.find('h3',class_='productBlock_productName')
    product_name = product_name.text.strip() if product_name else "N/A"

    # Extracting price
    price = product.find('span',class_='productBlock_priceValue')
    price = price.text if price else "N/A"

    # Extracting rating
    rating = product.find('span', class_='visually-hidden productBlock_rating_hiddenLabel')
    rating = rating.text if rating else "N/A"

    print(product_name)
    print(price)
    print(rating)

    skincare.append([product_name,price,rating]) # Append a list of product details

print(skincare)
```

```python
df=pd.DataFrame(skincare,columns=['product_name','price','review'])
df.head()
```

```python
# extract stars and reviews
df['stars'] = df['review'].str.extract(r'([\d\.]+) Stars')
df['review_counts'] = df['review'].str.extract(r'(\d+) Reviews')
df = df.drop('review', axis=1)
df.head(3)
```

```python
# define the rates
exchange_rate = 2  # 1 pound = 2 NZ dollar

# exchange the price to NZ currency
df['price'] = df['price'].str.replace('£', '').astype(float) * exchange_rate
df.head(3)
```

```python
# scale the stars
def convert_rating(stars):
    # Check if rating is a valid number before conversion
    if isinstance(stars, str) and stars.replace('.', '', 1).isdigit():  # Check if it's a string first
        return (float(stars) * 100) / 5
    else:
        return None  # Or handle non-numeric ratings as needed

# conver the rating
df['rating'] = df['stars'].apply(convert_rating)
df.drop('stars', axis=1, inplace=True)
# Swap columns
df = df[['product_name', 'price', 'rating', 'review_counts']]
df.head(3)
```

```python
df.dtypes
```

```python
df['review_counts'] = df['review_counts'].fillna(0).astype(int)
df.dtypes
```

```python
# Function to extract last four words from product_name
def extract_function(product_name):
    words = product_name.split()[-4:]
    return ' '.join(words)

# Apply the function to create 'function' column
df['function'] = df['product_name'].apply(extract_function)

# Split 'product_name' into 'product_name' and 'function' columns
df['name'] = df['product_name'].apply(lambda x: ' '.join(x.split()[:-4]))
df.head(5)
```

```python
df.drop('product_name', axis=1, inplace=True)
df.head(3)
```

```python
df.rename(columns={'name': 'product_name'}, inplace=True)
df.head(3)
```

```python
# Specify the desired column order
desired_columns = ['product_name','function','price','rating','review_counts']
# Reorder columns using reindex
df = df.reindex(columns=desired_columns)

df.head(3)
```
