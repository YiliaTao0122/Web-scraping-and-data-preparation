```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.sephora.nz/categories/skincare?view=120'
```

```python
respond = requests.get(url)
print(respond)
```

```python
response = respond.content
```

```python
soup= BeautifulSoup(response, 'html.parser')
soup
```

```python
table=soup.find('div',class_='products-container')
table
```

```python
products=table.find_all('div', class_='products-card-container')
products
```

```python
#find the rating list
rating = table.find_all('div', class_='stars')
rating
```

```python
#find the rating percentage element
rating_percentages = []

for div in rating:
    style_attribute = div.get('style')
    percentage = style_attribute.split('--highlighted-percentage:')[-1].split('%')[0]
    percentage_float = float(percentage) # Convert to float to handle decimals
    percentage_int = int(round(percentage_float)) # Round to nearest integer and then convert to int
    rating_percentages.append(percentage_int)

print(rating_percentages)
```

```python
#first approach, find all the review list
review=table.find_all('span', class_='reviews-count')
review
```

```python
#get the review element, signle element
review_counts = []

for span in review:  # Iterate over the review elements
    text = span.get_text(strip=True)
    review_counts.append(text)

print(review_counts)
```

```python
skincare=[]
for product in products:

    # Extracting product name
    product_name = product.find('div',class_='brand')
    product_name=product_name.text

    # Extracting price
    price=product.find('span',class_='sell-price')
    price=price.text

    # Extract function
    function=product.find('div',class_='product-name')
    function=function.text

    # Extract rating
    rating = product.find_all('div', class_='stars')
    highlighted_percentages = []
    for div in rating:
     style_attribute = div.get('style')
     percentage = style_attribute.split('--highlighted-percentage:')[-1].split('%')[0]
     percentage_float = float(percentage)
     percentage_int = int(percentage_float)  # Convert float to integer
     highlighted_percentages.append(percentage_int)
     rating=highlighted_percentages

    review=product.find_all('span', class_='reviews-count')
    review_counts = []
    for span in review:
        text = span.get_text(strip=True)
        review_counts.append(text)

    # Print extracted data
    print(product_name)
    print(price)
    print(function)
    print(rating)
    print(review_counts)

    skincare.append([product_name,function,price,rating,review_counts]) # Append a list of product details

print(skincare)
```

```python
# Convert the 'skincare' list (which contains lists of data) into a DataFrame
df = pd.DataFrame(skincare, columns=['product_name','function', 'price','rating','review_counts'])
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
# Remove dollar sign from price column
df['price'] = df['price'].str.replace('$', '')
# Convert price column to numeric
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

```python
# Extract numeric values from rating and review_counts
df['rating'] = df['rating'].apply(lambda x: int(x[0]) if isinstance(x, list) and len(x) > 0 else 0)  # Check if x is a non-empty list before indexing
df['review_counts'] = df['review_counts'].apply(lambda x: int(str(x[0]).replace('(', '').replace(')', '')) if isinstance(x, list) and len(x) > 0 else 0)  # Remove parentheses before converting to int
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Select numeric columns for outlier detection
numeric_columns = ['price', 'rating', 'review_counts']

# Calculate quartiles and IQR for numeric columns
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

# Determine outliers based on IQR for each numeric column
outliers = {}
for col in numeric_columns:
    lower_bound = Q1[col] - 1.5 * IQR[col]
    upper_bound = Q3[col] + 1.5 * IQR[col]
    outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

# Print outliers for each numeric column
for col, outlier_df in outliers.items():
    if not outlier_df.empty:
        print(f"Outliers in '{col}':")
        print(outlier_df)
        print()
```
